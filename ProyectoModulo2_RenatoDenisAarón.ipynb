# Enfriamiento en placas metalicas

- 1.1 El objetivo de este proyecto es **intentar conocer cual es el tiempo de enfriamiento, principalmente en las placas metalicas, con el uso de ecuaciones diferneciales** .

- 1.2 utlizando la ley del enfriamiento de Newton, para saber en cuanto tiempo se enfriaria una placa de metal, adotando su uso para la ingenieria civil. 


- # 1.3 Modelo

El modelo establece que la tasa de pérdida de calor de un cuerpo es proporcional a la diferencia de temperatura entre el cuerpo y sus alrededores con la ecuacion:

\begin{equation}
\frac{dT(t)}{dt}  = -k(T-Tm)
\end{equation}

Donde **T** es la tempretura del objeto, **Tm** es la temperatura del medio ambiente, y **k** es una constante de proporcionalidad.

- Para resolver esta ecuacion primero separamos las variables

\begin{equation}
\frac{dT(t)}{T-Tm}  = -k(dt)
\end{equation}

- Despues pasamos a intrgrar cada uno de sus lados

\begin{equation}
\int(\frac{dT(t)}{T-Tm})  = -k\int(dt)
\end{equation}

- Con lo cual obtenemos que la solucion de esta ecuacion diferencial es:

\begin{equation}
\ln({T-Tm})  = -kt+c
\end{equation}

- Y despejando

\begin{equation}
\ {T}  = Tm+Ce^{-kt}
\end{equation}


# 1.4 Simulacion

**Enfriamieto**

Tenemos una placa metalica, la cual  despues de soldarla para unir 2 piezas, se encuentra a una tempratura de 200 grados centigrados, el ambiente cuenta con una tempretura de 23 grados centigrados, y despues de 20 minutos esta a una tempratura de 100 grados queremos saber cuanto (tiempo en minutos) tardara eta placa en enfriarse para seguir trabajando, (25 grados seria la temperatura optima).

```python
import sympy as sym
sym.init_printing(use_latex='mathjax')
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
sym.var('t', real=True, positive=True)
T = sym.Function('T')
def func(T,k,t,x0,Tm):
    T1 = T[0]
    return np.array(-k*(T-Tm))
x0 = [200]
#asumiendo que en el minuto 20 habia una tempreatura de 100 grados
t = np.linspace(0,20)
Tm=23
k = 0.041617
T = odeint(func, x0, t, args=(T,k,Tm))
T1 = T[:,0]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, T1, 'r', lw=2, label='Temperatura')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutosx10]')
plt.ylabel('grados')
plt.show()
```

Como podemos observar en la grafica las condiciones mencionadas se cumplen, en t=20 la temperatura es igual a 100 grados, y podemos observar que alrededor del momento T=26 la tempreatura ya es optima para seguir trabajando.

# Ejemplo adicional (calentamiento)

Supongamos que se aplico nitrogeno liquido en una estructura, en ese momento la estructura se encuentra a una tempratura de -100 grados centigrados,en el minuto 20 la estructura tiene una temperatura de -60 grados, la teperatura ambiente es de 25 grados, ¿cuanto tiempo tardara la estructura en estar a una tempratura optima?

```python
sym.var('t', real=True, positive=True)
T = sym.Function('T')
def func(T,k,t,x0,Tm):
    T1 = T[0]
    return np.array(-k*(T-Tm))
x0 = [-100]
#asumiendo que en el minuto 20 habia una tempreatura de -60 grados
t = np.linspace(0,20)
Tm=25
k = -5.6961
T = odeint(func, x0, t, args=(T,k,Tm))
T1 = T[:,0]
```

```python
plt.figure(figsize=(8,6))
plt.plot(t, T1, 'r', lw=2, label='Temperatura')
plt.grid()
plt.legend(loc='best')
plt.xlabel('t [minutosx10]')
plt.ylabel('grados')
plt.show()
```

Como podemos ver esta estructura tendra una tempratura optima para trabajar aproximadamente al tiempo t=26.
